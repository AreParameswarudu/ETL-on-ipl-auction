```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.iplt20.com/auction'
page = requests.get(url)
html = BeautifulSoup(page.content,'html.parser')
teams = html.find_all('section',class_='ih-points-table-sec position-relative')

```

```python
title = []
player_name = []
base_bid = []
win_bid = []
cap=[]
category = []

for team in teams:
    team_name = team.find('h2').text
    table = team.find('table')
    trs = table.find_all('tr')
    for tr in trs[1:]:
          td = tr.find_all('td')
          title.append(team_name)
          player_name.append(tds[1].text)
          category.append('International' if tds[1].find('img') else 'National')
          base_bid.append(int(''.join(tds[2].text.split(',')))/10000000)
          win_bid.append(int(''.join(tds[3].text.split(',')))/10000000)
          cap.append(tds[4].text)


df = pd.DataFrame({
    'Team':title,
    'Player Name':player_name,
    'Category':category,
    'Base Bid':base_bid,
    'Win Bid':win_bid,
    'Cap':cap
})

df.to_csv('ipl_web_scraping.csv')
```
